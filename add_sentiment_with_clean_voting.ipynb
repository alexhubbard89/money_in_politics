```python
import pandas as pd
import numpy as np
```

```python
# load data
votes_df = pd.read_csv('cleaned_votes_money_history.csv')
votes_df['date'] = pd.to_datetime(votes_df['date'])
votes_df = votes_df.fillna(0)
bill_sentiment = pd.read_csv('bill_number_sentiment.csv')
```

# Add feature to describe if bill is more negative or positive. Higher numbers == more negative, where as 0 is not negative OR positive.

```python
bill_sentiment['n/p_ratio'] = bill_sentiment['Negative'] / bill_sentiment['Positive']
bill_sentiment['n/p_ratio'] = bill_sentiment['Negative'].replace('inf', 0)
bill_sentiment['n/p_ratio'] = bill_sentiment['Negative'].replace('nan', 0)
bill_sentiment.head(3)
```

```python
master = pd.DataFrame(columns=votes_df.columns)

for i in range(bill_sentiment.shape[0]):
    if i % 25 == 0:
        print i
    x = votes_df[votes_df['bill_id'] == bill_sentiment['Bill_number'][i]]
    x['Negative'] = bill_sentiment['Negative'][i]
    x['Positive'] = bill_sentiment['Positive'][i]
    x['Uncertainty'] = bill_sentiment['Uncertainty'][i]
    x['Litigious'] = bill_sentiment['Litigious'][i]
    x['Constraining'] = bill_sentiment['Constraining'][i]
    x['Superfluous'] = bill_sentiment['Superfluous'][i]
    x['Interesting'] = bill_sentiment['Interesting'][i]
    x['n/p_ratio'] = bill_sentiment['n/p_ratio'][i]
    master = master.append(x)
master = master.reset_index(drop=True)
master
```

```python
master.to_csv('cleaned_votes_money_history_with_sentiment.csv')
```
